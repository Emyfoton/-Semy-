```python
import numpy as np
import matplotlib.pyplot as plt


x= np.linspace(-10,10,100)
def f (x):
    return 3*x**5+5*x**4-x**3

def df (x):
    return 15*x**4+20*x**3-3*x**2


def roots (f,x0,max_iters=50,atol=1e-5,rtol=1e-5):
  for i in range (max_iters):
    r= x0 - f(x0)/df(x0)
    if np.isclose(0.,f(r),atol,rtol):
      return i,r
    x0=r
  return i,r

plt.plot(x,f(x), c= "pink")
plt.axhline(y=0, c='grey', linestyle=':')
plt.show()

n= roots(f,1,max_iters=50,atol=1e-5,rtol=1e-5)
n2= roots(f,-2,max_iters=50,atol=1e-5,rtol=1e-5)
print(n, n2)

```

```python
import numpy as np
import matplotlib.pyplot as plt

x= np.linspace(-10,10,100)
h= 0.01
def f (x):
    return 3*x**5+5*x**4-x**3

def df (f,x,h):
    d= 0
    if  h!=0:
      d= (f(x+h)-f(x-h))/(2*h)
    return d


def roots (f,x0, h, max_iters=50,atol=1e-5,rtol=1e-5):
  for i in range (max_iters):
    r= x0 - f(x0)/df(f, x0, h)
    if np.isclose(0.,f(r),atol,rtol):
      return r
    x0=r
  return r

plt.plot(x,f(x), c= "pink")
plt.axhline(y=0, c='grey', linestyle=':')
plt.show()

n= roots(f,1, h, max_iters=50,atol=1e-5,rtol=1e-5)
n2= roots(f,-2, h, max_iters=50,atol=1e-5,rtol=1e-5)
n3= roots(f,0, h, max_iters=50,atol=1e-5,rtol=1e-5)
print(n, n2, n3)
```
