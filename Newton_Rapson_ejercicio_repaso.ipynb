```python
import numpy as np
import matplotlib.pyplot as plt


x= np.linspace(-10,10,100)
def f (x):
    return 3*x**5+5*x**4-x**3

def df (x):
    return 15*x**4+20*x**3-3*x**2


def roots (f,x0,max_iters=50,atol=1e-5,rtol=1e-5):
  for i in range (max_iters):
    r= x0 - f(x0)/df(x0)
    if np.isclose(0.,f(r),atol,rtol):
      return i,r
    x0=r
  return i,r

plt.plot(x,f(x))
plt.axhline(y=0, c='r', linestyle='--')
plt.show()

n= roots(f,0.2,max_iters=50,atol=1e-5,rtol=1e-5)
print(n)

```
